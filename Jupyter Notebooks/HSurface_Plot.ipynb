```python
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
```

```python
init_conc = np.linspace(0.25, 100, 112)
x = []
y = []
for file in glob.glob('../../MCMC_Data/mcmc_unique_*'):
    idx = int(os.path.basename(file).split('_')[-1].split('.')[0])
    x.append(init_conc[idx])
    data = np.loadtxt(file)
    y.append(data.mean())

x = np.array(x)
y = np.array(y)
```

```python
perfect = -8.949*8400

surface = -73934.85120670185

surface_area = 54.45950522225023*51.344913923623125

N_h = np.ceil(400*x*1e-2)

surface_energy = (surface - y*N_h - perfect)/ (2*surface_area ) 

plt.scatter(x,surface_energy)
plt.title('Surface Energy of 111 Surface as a function of Hydrogen Coverage')
plt.ylabel('Surface Energy ev/A^2')
plt.xlabel('Hydrogen Coverage / %')
```
