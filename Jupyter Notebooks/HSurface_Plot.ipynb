```python
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
```

```python
init_conc = np.linspace(0.25, 100, 112)
x = []
y = []
for file in glob.glob('../MCMC_Data/mcmc_unique_*'):
    idx = int(os.path.basename(file).split('_')[-1].split('.')[0])
    x.append(init_conc[idx])
    data = np.loadtxt(file)
    y.append(data.max())
    print(data.var())

x = np.array(x)
y = np.array(y)
```

```python
plt.scatter(x,x*y)
plt.ylabel('Binding Energy ev/%')
plt.xlabel('Hydrogen Coverage / %')
```
